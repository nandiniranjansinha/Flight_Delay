```python
jan = pd.read_csv('Jan_2025.csv')
feb = pd.read_csv('Feb_2025.csv')
mar = pd.read_csv('Mar_2025.csv')
apr = pd.read_csv('Apr_2025.csv')
```

```python
# Combining the csv files
df = pd.concat([jan,feb,mar,apr], ignore_index = True)
# ignore_index to remove the index (unnecessary column)
```

```python
# Top 5 airlines by number of flights
df['OP_UNIQUE_CARRIER'].value_counts().head()
```

```python
# Average arrival delay by the airline
df.groupby('OP_UNIQUE_CARRIER').ARR_DELAY.mean().head()
```

```python
# We have not sorted anything here yet, so it just shows randomish results by just grouping the airlines
```

```python
# To show in descending order
df.groupby("OP_UNIQUE_CARRIER").ARR_DELAY.mean().sort_values(ascending=False).head()
```

OO is the least delayed - arrival

OH is the most delayed - arrival

```python
# Let's find the top busiest airports
df['ORIGIN'].value_counts().head()
```

```python
# Let's save the df to a csv file for future data use
df.to_csv('Flight_Delay_JanApr2025.csv', index = False)
```
