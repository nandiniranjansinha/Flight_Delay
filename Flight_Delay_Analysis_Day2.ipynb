Since, the value is right-skewed for DEP_DELAY, we must go with median. But only median for this column may not give the right result considering different airlines. Like airline "A" can be 5 mins late whereas airline "B" can be 30 mins late. We never know. So grouping them w.r.t the airline would give more accurate result.

```python
DF['DEP_DELAY'].isnull().sum()
```

```python
DF['DEP_DELAY'] = DF.groupby('OP_UNIQUE_CARRIER')['DEP_DELAY'].transform(lambda x:x.fillna(x.median()))
```

```python
DF['DEP_DELAY'].isnull().sum()
```

```python
DF['ARR_DELAY'].isnull().sum()
```

```python
DF['ARR_DELAY'] = DF.groupby('OP_UNIQUE_CARRIER')['ARR_DELAY'].transform(lambda x:x.fillna(x.median()))
```

```python
delay_new = ['DEP_DELAY_NEW','ARR_DELAY_NEW']

for new in delay_new:
  DF[new] = DF.groupby('OP_UNIQUE_CARRIER')[new].transform(lambda x:x.fillna(x.median()))
```

```python
DF['ARR_DELAY_NEW'].isnull().sum()
```

```python
DF.isnull().sum()
```

x.mode() returns the most frequent value.

.mode()[0] grabs the first mode, in case of ties.

if not x.mode().empty: checks that mode exists.

If x.mode() is empty (edge case), it fills with 0 as a fallback.

```python
delay_15 = ['DEP_DEL15','ARR_DEL15']

for delay in delay_15:
  DF[delay] = DF.groupby('OP_UNIQUE_CARRIER')[delay].transform(lambda x:x.fillna(x.mode()[0] if not x.mode().empty else 0))
```

# 📊 Exploratory Data Analysis (EDA) – Flight Delay Dataset

### 🔍 What is EDA?

Exploratory Data Analysis (EDA) is the process of understanding and analyzing the dataset before building any model.
It helps us:

* Detect **patterns and trends**
* Identify **missing values and outliers**
* Understand the **distribution of features**
* Decide how to **clean and prepare data** for modeling

---

## 1️⃣ Delay Features (Boxplots)

* **Departure Delay (DEP\_DELAY & DEP\_DELAY\_NEW)**

  * Most flights depart on time or with only small delays.
  * Some flights experience **very large delays** (outliers visible as long whiskers).
  * A few flights depart earlier than scheduled (negative delays).
  * Distribution is **right-skewed** → mean is pulled up by extreme delays.

* **Arrival Delay (ARR\_DELAY & ARR\_DELAY\_NEW)**

  * Pattern is similar to departure delay.
  * Majority are close to schedule, but **a subset of flights face major delays**.
  * Some negative values represent **early arrivals**, but they are rare.

* **Binary Delay Columns (DEP\_DEL15 & ARR\_DEL15)**

  * These indicate whether delay ≥ 15 minutes (1) or not (0).
  * Boxplots don’t show much since they are binary, but they will be useful for classification later.

**Insight:**
Most flights run close to schedule, but a small portion of flights suffer **very large delays**. Using **median** is more reliable than mean for these skewed distributions.

---

## 2️⃣ Flight Duration & Distance (Histograms)

* **Actual Elapsed Time & CRS Elapsed Time**

  * Most flights last between **50–200 minutes**.
  * Few flights are long-haul (> 300 mins).
  * Distribution is **right-skewed** → short flights dominate the dataset.

* **Distance**

  * Majority of flights cover **short to medium distances (< 1000 miles)**.
  * Very few flights travel beyond 2000 miles.

**Insight:**
Dataset is dominated by **short to medium-haul flights**, which may affect delay patterns differently compared to long-haul flights.

---

## 3️⃣ Delay Distributions (Histograms)

* **Departure Delay**

  * Most flights are clustered around **0–20 mins delay**.
  * Some flights have extreme delays (> 100 mins).
  * A small number of flights leave earlier than scheduled (negative values).

* **Arrival Delay**

  * Similar to departure delays, but slightly more spread out.
  * A subset of flights shows **very large arrival delays (> 150 mins)**.

**Insight:**
Both departure and arrival delays are **heavily right-skewed** with a long tail of extreme delays.

---

## 4️⃣ Airline-Level Analysis

* **Average Departure Delay by Airline**

  * Some airlines consistently show **higher average delays** compared to others.
  * Indicates differences in **operational efficiency**.

* **Flight Count per Airline**

  * A few airlines dominate the dataset with much higher number of flights.
  * Important: More flights could naturally lead to more delays in total.

**Insight:**
Airline performance varies significantly. Some carriers manage punctuality better, while others face frequent delays.

---

## 5️⃣ Airport-Level Analysis

* **Top 10 Origin Airports (Departure Delays)**

  * Certain airports stand out with **very high departure delays**.
  * Likely due to congestion, weather conditions, or operational bottlenecks.

* **Top 10 Destination Airports (Arrival Delays)**

  * Similarly, some destinations consistently record higher arrival delays.
  * Suggests airport infrastructure and weather significantly affect punctuality.

**Insight:**
Airports play a critical role in delays – both at **origin and destination**.

---

## 6️⃣ Flight Status: Cancelled & Diverted

* **Overall Trends**

  * Cancelled flights are **much more frequent** than diverted flights.

* **Per Airline**

  * Some airlines have higher counts of cancellations/diversions.
  * This may be linked to route type, fleet management, or weather exposure.

**Insight:**
Cancellations are a bigger issue than diversions, and their distribution varies by airline.

---

## 7️⃣ Flight Distance vs Arrival Delay (Scatterplot)

* No strong correlation between **flight distance** and **arrival delay**.
* Delays occur across both short and long flights.
* Suggests that **distance itself is not a direct factor** for delays.

**Insight:**
Factors like weather, airline operations, and airport congestion are more important than distance in determining delays.

---

# 📝 Final Key Takeaways

1. **Most flights are on time**, but a small subset experience **very large delays**, making the data right-skewed.
2. **Short and medium-haul flights dominate** the dataset.
3. **Departure and arrival delays are correlated** – late departures usually mean late arrivals.
4. **Airline performance varies** – some carriers manage punctuality better.
5. **Airports matter** – congestion/weather at certain hubs cause frequent delays.
6. **Cancellations are more common** than diversions, with differences across airlines.
7. **Distance is not a major driver of delays** – delays happen regardless of flight length.gi
